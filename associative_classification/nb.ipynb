```python
a = [{1, 2, 4}, {2, 3, 1}, {5,1, 2}]
from itertools import combinations
candidate = set()
for i in range(0, len(a)):
    *first, last = sorted(a[i])
    tail = [last]
    for j in range(0, len(a)):
        if a[i] == a[j]:
            continue
        *f, l = sorted(a[j])

        if first == f:
            tail.append(l)
    
    for combination in combinations(tail, 2):
        candidate.add(frozenset(sorted(first + list(combination))))


candidate
```

```python
from pyarc import TransactionDB
from pyarc.algorithms import (
    top_rules,
    createCARs,
    M1Algorithm,
    M2Algorithm
)
import pandas as pd

from sklearn.model_selection import train_test_split

nursery_data = "/home/haianh/grad_project/ml-learning/data/nursery_pyarc.data"

data = pd.read_csv(nursery_data)

data_train, data_test = train_test_split(data, test_size=0.2, shuffle=True)

txns_train = TransactionDB.from_DataFrame(data_train)
txns_test = TransactionDB.from_DataFrame(data_test)

# get the best association rules
rules = top_rules(txns_train.string_representation)

# convert them to class association rules
cars = createCARs(rules)

classifier = M1Algorithm(cars, txns_train).build()
# classifier = M2Algorithm(cars, txns_train).build()

accuracy = classifier.test_transactions(txns_test)
```

```python
class Rule:
    def __init__(self, cond_set, class_label) -> None:
        self.cond_set: tuple[tuple] = cond_set
        self.class_label = class_label

    def __repr__(self) -> str:
        return f"{self.cond_set} --> {self.class_label}"

def cond_set(rule):
    return rule.cond_set

r1 = Rule(((1, 2), (3, 5)), 1)
r2 = Rule(((2, 3), (1, 5)), 2)

l = []

l.append(r2)
l.append(r1)

nl = sorted(l, key=cond_set)

nl
```
