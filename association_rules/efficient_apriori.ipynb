Benchmark of apriori in diff datasets

```python
import csv
from efficient_apriori import apriori
from time import perf_counter

SUPPORT_LIST = [0.05, 0.03, 0.01]
CONFIDENCE_LIST = [0.5, 0.7, 0.9]

def read_data(data_path):
    data = []
    with open(data_path, 'r') as data_file:
        reader = csv.reader(data_file)
        for line in reader:
            data.append(line)
    return data

def count_itemset(freq_itemsets):
    num = 0
    for k in freq_itemsets:
        num += len(freq_itemsets[k])
    return num


def bench(transactions, support_list: list, confidence_list: list):
    for support in support_list:
        for confidence in confidence_list:
            start = perf_counter()
            frequent_itemsets, rules = apriori(transactions, min_support=support, min_confidence=confidence)
            end = perf_counter()
            num_itemset = count_itemset(frequent_itemsets)
            num_rules = len(rules)
            print(f"Support = {support}, Confidence = {confidence} --> Time: {end-start} seconds; Number of itemsets: {num_itemset}, Number of rules: {num_rules}")

```

```python
grocery_data_path = "/home/haianh/grad_project/ml-learning/association-rule-mining-apriori/data/groceries.csv"
adult_data_path = "/home/haianh/grad_project/ml-learning/data/adult.data"
letter_data_path = "/home/haianh/grad_project/ml-learning/data/letter-recognition.data"
```
