```python
nursery_data_path = "/home/haianh/grad_project/ml-learning/data/nursery.data"
letter_data_path = "/home/haianh/grad_project/ml-learning/data/letter-recognition.data"
```

```python
data_train, data_test = train_test_split(pd.read_csv(nursery_data_path),test_size=0.33)
```

```python
supports = [0.1, 0.05, 0.03, 0.02, 0.01, 0.005]
```

```python
from time import perf_counter

execution_time = []
accuracy_list = []

for sup in supports:
    start = perf_counter()

    txns_train = TransactionDB.from_DataFrame(data_train)
    txns_test = TransactionDB.from_DataFrame(data_test)
    cba = CBA(support=sup, confidence=0.5, algorithm="m1")
    cba.fit(txns_train)
    
    accuracy = cba.rule_model_accuracy(txns_test) 
    end = perf_counter()
    execution_time.append(end-start)
    accuracy_list.append(accuracy)
```
