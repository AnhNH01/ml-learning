```python
def map_categorical(data, column_no):
    """
    Map categorical value with a positive integer
    
    Eg: ('red', 'green', 'blue') --> (1, 2, 3)
    """
    size = len(data)
    classes = set([x[column_no] for x in data])
    classes_no = dict([(label, 0) for label in classes])
    j = 1
    for i in classes:
        classes_no[i] = j
        j += 1
    for i in range(size):
        data[i][column_no] = classes_no[data[i][column_no]]
    return data
```

```python

import csv

def read_data(path):
    data = []
    with open(path, 'r') as csv_file:
        reader = csv.reader(csv_file, delimiter=',')
        for line in reader:
            data.append(line)
        while [] in data:
            data.remove([])
    return data


def process_data(data):
    for i in range(0, 8):
        data = map_categorical(data, i)
    return data

data = process_data(read_data(nursery_data_path))

data
```
